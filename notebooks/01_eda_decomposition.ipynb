# Exploratory Analysis & Seasonality Decomposition

Inspecting trend, seasonality, and patterns in the retail sales series before modeling.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from src.data.load import load_raw, filter_series

plt.rcParams['figure.figsize'] = (12, 5)
```

## 1. Load a Representative Series

```python
raw = load_raw()
series = filter_series(raw, store=1, item=1)
series.set_index('date')['sales'].plot(title='Daily Sales — Store 1, Item 1')
plt.ylabel('Units Sold')
plt.show()
```

## 2. Seasonal Decomposition

Splitting the series into trend, seasonal, and residual components reveals the underlying structure Prophet will try to model explicitly.

```python
ts = series.set_index('date')['sales']
decomposition = seasonal_decompose(ts, model='additive', period=365)

fig, axes = plt.subplots(4, 1, figsize=(12, 10), sharex=True)
decomposition.observed.plot(ax=axes[0], title='Observed')
decomposition.trend.plot(ax=axes[1], title='Trend')
decomposition.seasonal.plot(ax=axes[2], title='Seasonal')
decomposition.resid.plot(ax=axes[3], title='Residual')
plt.tight_layout()
plt.show()
```

## 3. Day-of-Week Pattern

```python
series['day_of_week'] = series['date'].dt.day_name()
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
avg_by_dow = series.groupby('day_of_week')['sales'].mean().reindex(order)
avg_by_dow.plot(kind='bar', color='steelblue', title='Average Sales by Day of Week')
plt.ylabel('Avg Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. Monthly Seasonality

```python
series['month'] = series['date'].dt.month
avg_by_month = series.groupby('month')['sales'].mean()
avg_by_month.plot(kind='bar', color='coral', title='Average Sales by Month')
plt.ylabel('Avg Units Sold')
plt.tight_layout()
plt.show()
```
